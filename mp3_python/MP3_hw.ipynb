````python
def get_mic_signal(signals):
    return signals[0],signals[1],signals[2],signals[3],signals[4],signals[5]
````

````python
def sample_lag(sample_a,sample_b):
    correlation = signal.correlate(sample_a, sample_b, mode="full")
    correlation_lags = signal.correlation_lags(sample_a.size, sample_b.size, mode="full")
    sample_lag = correlation_lags[np.argmax(correlation)]
    return sample_lag
````

````python
def aoa_localization(mic_data_folder, FS, MIC_OFFSETS):
    """AoA localization algorithm. Write your code here.

    Args:
        mic_data_folder: name of folder (without a trailing slash) containing 
                         the mic datafiles `0.csv` through `7.csv` and `config.csv`.
        FS: microphone sampling frequency - 16kHz.
        MIC_OFFSETS: a list of tuples of each microphone's location relative to the center of its mic array. 
                     This list is calculated based on the diameter(0.09218m) and geometry of the microphone array.
                     For example, MIC_OFFSETS[1] is [0.09218*0.5, 0]. If the location of microphone array #i is
                     [x_i, y_i]. Then [x_i, y_i] + MIC_OFFSETS[j] yields the absolute location of mic#j of array#i.
                     This is provided for your convenience and you may choose to ignore.

    Returns:
        The user's location in this format (in metres): (L_x, L_y)

    """
    list_val = []

    for mic_num in np.arange(8):
        _ = pd.read_csv("{}/{}.csv".format(mic_data_folder, mic_num),header=None)
        list_of_mic_signals = get_mic_signal(_)
        zero_mic,mic_1,mic_2,mic_3,mic_4,mic_5 = get_mic_signal(_)

        list_sample_lag_mic_0_to_rest = []

        total_dist = 0
        for mic in range(0,6):
            num_sample_lag = sample_lag(list_of_mic_signals[0],list_of_mic_signals[mic])
            time_lag = num_sample_lag/FS
            dist = float(343)*time_lag
            print(float(dist))
#             total_dist += abs(float(dist))
            total_dist += (float(dist))
        total_dist= abs(total_dist)
        list_val.append(total_dist)
        print("=====================")
    idx_nearest_ap = list_val.index(max(list_val))
#     idx_nearest_ap = 4
    print("This is mic array that is nearest to current position",idx_nearest_ap)
    _ = pd.read_csv("{}/{}.csv".format(mic_data_folder, idx_nearest_ap),header=None)
    list_of_mic_signals = get_mic_signal(_)
    list_of_dist_abs = []
    list_of_dist = []
    for mic in range(0,6):
        num_sample_lag = sample_lag(list_of_mic_signals[0],list_of_mic_signals[mic])
        time_lag = num_sample_lag/16000
        dist = float(343)*time_lag
        list_of_dist.append(dist)
        list_of_dist_abs.append(abs(dist))
#         print(float(dist))
    list_of_furthest_idx = []
    largest_value = max(list_of_dist_abs)
    for idx, value in enumerate(list_of_dist):
        if largest_value == abs(value):
            list_of_furthest_idx.append(idx)
    print(list_of_furthest_idx)
    
    furthest_dist_idx = list_of_dist_abs.index(max(list_of_dist_abs))
    

    dici = {}
    dici[1] = 2*math.pi/3
    dici[2] = math.pi/2
    dici[3] = math.pi/3
    dici[4] = math.pi/6
    dici[5] = 0
#     print("This is furthest",furthest_dist_idx)

    speaker_pos = pd.read_csv("{}/{}.csv".format(mic_data_folder, "config"),header=None)
    list_speaker_position = speaker_pos.values.tolist()
    
    for index in list_of_furthest_idx:
#         print(index)
        print("This is the position",list_speaker_position[idx_nearest_ap])
        if list_of_dist[index] < 0:
            list_speaker_position[idx_nearest_ap][0] += math.cos(dici[index])/len(list_of_furthest_idx)
            list_speaker_position[idx_nearest_ap][1] += math.sin(dici[index])/len(list_of_furthest_idx)
            print("this is what we are adding",math.cos(dici[index])/len(list_of_furthest_idx),math.sin(dici[index])/len(list_of_furthest_idx))
        else:
            list_speaker_position[idx_nearest_ap][0] -= math.cos(dici[index])/len(list_of_furthest_idx)
            list_speaker_position[idx_nearest_ap][1] -= math.sin(dici[index])/len(list_of_furthest_idx)
            print("this is what we are adding",-math.cos(dici[index])/len(list_of_furthest_idx),-math.sin(dici[index])/len(list_of_furthest_idx))
    return (list_speaker_position[idx_nearest_ap][0],list_speaker_position[idx_nearest_ap][1])


    # Your return value should be the user's location in this format (in metres): (L_x, L_y)
    return (0.0, 1.0) 
````
